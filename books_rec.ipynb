```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from sklearn.preprocessing import MinMaxScaler
import pickle
```

```python
books_df = pd.read_csv('books_final.csv')
```

```python
books_df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MinMaxScaler

# Combine Genre and Author into a new feature
books_df['Genre_Author'] = books_df['Genre']+' '+books_df['Author']

# Use TF-IDF Vectorizer to convert the combined text data into numerical features
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(books_df['Genre_Author'])

# Scale the ratings between 0 and 1
scaler = MinMaxScaler()
books_df['Scaled_Rating'] = scaler.fit_transform(books_df[['Rating']])
```

```python
print(tfidf_matrix[0])
```

```python
books_df.head()
```

```python
from sklearn.metrics.pairwise import linear_kernel

# Compute the cosine similarity matrix for TF-IDF features
cosine_sim_text = linear_kernel(tfidf_matrix, tfidf_matrix)

# Compute the cosine similarity matrix for ratings
rating_matrix = books_df[['Scaled_Rating']].values
cosine_sim_rating = linear_kernel(rating_matrix, rating_matrix)

# Combine the similarities: weighted sum
combined_sim = 0.7 * cosine_sim_text + 0.3 * cosine_sim_rating
```

```python
with open('tfidf_matrix.pkl', 'wb') as file:
    pickle.dump(tfidf_matrix, file)

with open('combined_sim.pkl', 'wb') as file:
    pickle.dump(combined_sim, file)

with open('books_df.pkl', 'wb') as file:
    pickle.dump(books_df, file)
```
